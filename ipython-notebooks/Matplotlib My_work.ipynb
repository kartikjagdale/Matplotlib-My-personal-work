##Example
A simple figure with MATLAB like plotting API:

```python
x = linspace(0,5,10)
y = x ** 2
```

```python
figure()
plot(x,y,'r')
xlabel('X')
ylabel('Y')
title('Simple Graph')
show()
close()
```

You can also subplot and color/symbol selection

```python
subplot(1,2,1)
title('Subplot 1')
xlabel('X')
ylabel('Y')
plot(x,y, 'r--')
subplot(1,2,2)
plot(y, x, 'g*-');
title('Subplot 2')
xlabel('X')
ylabel('Y')
show()
close()
```

##The Matplotlib Object-Oriented API

```python
#Create an instance of Global Figure
fig = figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height (range 0  to 1)
axes.plot(x, y, 'c*-')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('GRAPH using OOP API')
show()
close()
```

A little bit code is involved, but the advantage is we have full control over the plot axes are placed and we can easily add more than one axis to the figure

```python
fig = plt.figure(figsize=(10,8), dpi=100)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
axes3 = fig.add_axes([0.6, 0.2, 0.29, 0.19]) # inset axes

# main Figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Main Title')

# inset 1
axes2.plot(y, x, 'g*--')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Inset Title 1')

# inset 2
axes3.plot(x, y, 'x--')
axes3.set_xlabel('X')
axes3.set_ylabel('Y')
axes3.set_title('Inset Title 2')
show()
close()
```

###Legends, Labels and Titles

```python
fig, axes = plt.subplots(figsize=(12,6))

axes.plot(x *2, y*2, 'r*-',label='curve1')
axes.plot(x , y, 'b--',label='curve2')
axes.plot(y*2 , x*3, 'g*-',label='curve3')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Figure with Size Defined')
axes.legend(['curve1','curve2','curve3'])
fig.savefig('withlegend.png', dpi=300)
show()
close()
```

The ***legend*** takes optional keyword argument *loc* that can be used to specify the location of the figure is to be drawn

```python
axes.legend(loc=0) #let matplotlib decide the optimal location
axes.legend(loc=1) #upper right corner
axes.legend(loc=2) #upper left corner
axes.legend(loc=3) #lower left corner
axes.legend(loc=4) #lower right corner
# ... many more options are availabel
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2); #upper left corner
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Changing the Location of legend');
show()
close()
```

##Formating text: LaTeX, fontsize, font family
For example "$y=x^3$"

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, x**3, 'r--',label=r"$y=\alpha^3$")
ax.plot(x, x**2, 'b*-', label=r"$y= \alpha^2$")
ax.plot(x, x**2.3, 'c*-', label=r"$y= \alpha^2.3$")
ax.legend(loc=2) #upper lef corner
ax.set_xlabel(r"$X$", fontsize=18)
ax.set_ylabel(r"$Y$",fontsize=18)
ax.set_title(r'$LaTeX$')
fig.savefig("LaTeX.png", dpi=200)
show()
close()
```

We can also chaneg the gloabal font by executing following statements

```python
# Uodate the matplotlib config parameters:
matplotlib.rcParams.update({'font.size':18, 'font.family':'serif'})
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(x, x**3, 'r--',label=r"$y=\alpha^3$")
ax.plot(x, x**2, 'b*-', label=r"$y= \alpha^2$")
ax.plot(x, x**2.3, 'c*-', label=r"$y= \alpha^2.3$")
ax.legend(loc=2) #upper lef corner
ax.set_xlabel(r"$X$", fontsize=18)
ax.set_ylabel(r"$Y$",fontsize=18)
ax.set_title(r'$LaTeX$_with Serif Font')
fig.savefig("LaTeX_with serif_font.png", dpi=200)
show()
close()
```

###We can also change the color

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the color and alpha keyword arguments:

```python
fig, ax = plt.subplots()
ax.plot(x, x+1, color = "red", alpha=0.5) # half transperent red
ax.plot(x, x+2, color = "#1155dd") # RGD hex code for a bluish color
ax.plot(x, x+3, color = "#15cc55") # RGD hex code for a greenish 
show();close();
```

###Line and Marker Styles
to change the the *line* width, we can use the *linewidth* or *lw* keyworkd arg. and the Linestyle can be change using the *linestyle*  or *ls* keyword arg.

```python
matplotlib.rcParams.update({'font.size':12})
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="blue", linewidth=0.25)
ax.plot(x, x+2, color="blue", lw=0.50) # using the lw keyword
ax.plot(x, x+3, color="blue", linewidth=1.00)
ax.plot(x, x+4, color="blue", linewidth=2.00)

# possible linestyles options = '- ' '-' '-.' ':' 'steps'

ax.plot(x, x+5, color="red", lw=2, linestyle ='-')
ax.plot(x, x+6, color="red", lw=2,ls='-.') # using the lw keyword
ax.plot(x, x+7, color="red", lw=2,ls=':')
ax.plot(x, x+8, color="red", lw=2,ls='steps')

# also manay more params are ther like'
# marker symbols:= '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...

# marker size = markersize and marker color = markerfacecolor

# markeredgewidth and markeredgecolor

show();close();
```

###Control over axis and appearance

###Plot Range

we can configure the ranges of the axes using *ser_ylim* andset_ xlim or axis('tight') for auto getting "titly fitted" axes ranges.

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].plot(x,x**2,x, x**3)
axes[0].set_title("Default axes ranges")

axes[1].plot(x,x**2,x, x**3)
axes[1].axis('Tight')
axes[1].set_title("Tight axes")

axes[2].plot(x,x**2,x, x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title("Custom axes range")
show();close();
```

###Logarithmic Scale

Custom Scale is possible using *set_xscale* and *set_yscale* which accept one parameter (with the value "log" in this case)

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Normal Scale")

axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log") # see y axes to see the change
axes[1].set_title("Logarithmic scale (y)");
show(); close();
```

##Placement of ticks and custom tick labels

we can explictly determine where we cant the axis with *set_xticks* and *set_yticks* and we can also use *set_xticklabels* and *set_yticklabels*

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, x**2,'g*--' ,x, x**3,'r--', lw=2, )

ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$',r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0,50,100,150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$"% y for y in yticks], fontsize=18); # use LaTeX formated labels
show(); close();
```
